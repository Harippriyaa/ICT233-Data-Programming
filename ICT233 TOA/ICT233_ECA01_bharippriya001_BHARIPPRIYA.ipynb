## Qn 1a

```python
#load datasets
# copy df to a new variable to avoid accidental changes in original data
import pandas as pd

humidity = pd.read_csv('humidity.csv')
humidity_new = humidity.copy()

rain = pd.read_csv('rain.csv')
rain_new = rain.copy()

temperature = pd.read_csv('temperature.csv')
temperature_new = temperature.copy()

wind = pd.read_csv('wind.csv')
wind_new = wind.copy()

others = pd.read_csv('others.csv')
others_new = others.copy()
```

```python
#Step 1 - create new column called 'Sensor Type' for each dataframe

humidity_sensor = "humidity"
humidity_new['Sensor Type'] = humidity_sensor

rain_sensor = "rain"
rain_new['Sensor Type'] = rain_sensor

temperature_sensor = "temperature"
temperature_new['Sensor Type'] = temperature_sensor

wind_sensor = "wind"
wind_new['Sensor Type'] = wind_sensor

others_sensor = "others"
others_new['Sensor Type'] = others_sensor
```

```python
#Step 2 - Combine 5 dataframes into ONE (1) dataframe and sort it by Measurement Timestamp ascendingly

merged_dataframe = pd.concat([humidity_new, temperature_new, rain_new, wind_new, others_new])
merged_dataframe.sort_values("Measurement Timestamp", axis = 0, ascending = True,inplace = True, ignore_index=True)

display(merged_dataframe)
```

## Qn 1b

```python
new = merged_dataframe.copy()

new.update(new.groupby("Station Name").ffill())
new.sort_values(["Station Name","Measurement Timestamp"], axis = 0, ascending = True,inplace = True, ignore_index=True)

display(new)
```

## Qn 1c

```python
new_dataframe = new.copy()
new_dataframe = new_dataframe.drop('Sensor Type',axis=1)

new_dataframe['Hourly Timestamp'] = new_dataframe['Measurement Timestamp']
new_dataframe['Hourly Timestamp'] = pd.to_datetime(new_dataframe['Hourly Timestamp']).dt.round('h')

new_dataframe = new_dataframe.groupby('Station Name').resample('60min', on='Hourly Timestamp', loffset='15min59s').last().reset_index(drop=True)

new_dataframe = new_dataframe[['Station Name','Hourly Timestamp','Measurement Timestamp','Humidity','Air Temperature','Wet Bulb Temperature','Rain Intensity','Interval Rain','Total Rain','Wind Direction','Wind Speed','Maximum Wind Speed','Heading','Precipitation Type','Barometric Pressure','Solar Radiation','Battery Life']]
display(new_dataframe)
```

## Qn 1d

```python
%matplotlib inline
import matplotlib.pyplot as plt

graph_df = new_dataframe.copy()
graph_df['Hourly Timestamp'] = graph_df['Hourly Timestamp'].astype(str)
fig,ax = plt.subplots(figsize=(30,30))
plt.xlabel('Hourly Tmestamp')
plt.ylabel('Air Temperature')

plt.legend(loc=50, prop={'size': 50})
graph_df.groupby(['Hourly Timestamp','Station Name']).mean()['Air Temperature'].unstack().plot(ax=ax)
```

First Insight
The air temperature decreases as time decreases among all the three stations.

Second Insight
Only the temperature in Foster Weather Station went till -10 as compared to other stations.

```python
import requests
from bs4 import BeautifulSoup
import sqlite3

user_agent = {'User-agent': 'Mozilla/5.0'}

BASE_URL = 'http://en.wikipedia.org'
page = requests.get(BASE_URL + '/w/index.php?title=Family_tree_of_British_monarchs&oldid=1043575587', headers = user_agent)

soup = BeautifulSoup(page.content)
table = soup.find('table')


names_and_links = []
names = []
links = []

for tr in table.find_all('tr'):
    for td in tr:
        if td.find('a'):
            if td.find('span'):
                names_and_links.append(td)
    

for td in names_and_links:
    name_link = td.find('a')
    name = name_link['title']
    if 'House' not in name:
        names.append(name)
            
    if name_link:
        if 'House' not in name_link.attrs['href']:
            links.append(name_link.attrs['href'])

```

## Qn 2f

```python
import sqlite3

conn = sqlite3.connect('BritishMonarchsDB.sqlite')
cur = conn.cursor()

descendants = []
counter = 0

cur.execute('''SELECT id FROM british_monarch_family_tree WHERE name LIKE '%Queen Victoria%' ''')
record = cur.fetchone()

cur.execute('SELECT * FROM british_monarch_family_tree WHERE mother_id = (?) OR father_id = (?)', (record[0],record[0],))
descendantiD = cur.fetchall()
for descendant_id in descendantiD:
    descendants.append(descendant_id)

while len(descendants) != counter:
    counter += 1
    cur.execute('SELECT * FROM british_monarch_family_tree WHERE mother_id = (?) OR father_id = (?)',(descendants[counter-1][0],descendants[counter-1][0],))
    descendantiD = cur.fetchall()
    for descendant_id in descendantiD:
        descendants.append(descendant_id)

print(f'The descendants of Queen Victoria are: ')
for descendant in descendants:
    print (f'{descendant[1]}')
```
